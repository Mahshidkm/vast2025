```python
from helper import *
import json

# Define all possible work-related roles
PERSON_ROLES = ['PerformerOf', 'ComposerOf', 'ProducerOf', 'LyricistOf']

def prepare_data_for_career_of_artists(G):
    data = []
    for node, attrs in G.nodes(data=True):
        if attrs.get('Node Type') == 'Person':
            artist_name = attrs.get('name')
            artist_works, _ = find_all_artist_works(G, node)   # returns {work: set(roles)}
            collaborators = find_collaborators_of_artist(G, node)
            influenced_artists = find_directly_influenced(G, node)

            years = []
            distinct_genres = {}
            role_counts = {role: 0 for role in PERSON_ROLES}
            oceanus_yearly = {}   # new: year -> count of Oceanus Folk works

            # Count roles and collect Oceanus works per year
            for work, roles in artist_works.items():
                for role in roles:
                    if role in role_counts:
                        role_counts[role] += 1
                # Check if this work is Oceanus Folk
                work_genre = G.nodes[work].get('genre')
                if work_genre == 'Oceanus Folk':
                    release_date = G.nodes[work].get('release_date')
                    if release_date:
                        year = int(release_date)
                        oceanus_yearly[year] = oceanus_yearly.get(year, 0) + 1

            # ---- Yearly and genre info (as before) ----
            for artist_work in artist_works:
                release_date = G.nodes[artist_work].get('release_date')
                genre = G.nodes[artist_work].get('genre')
                if release_date:
                    years.append(int(release_date))
                if genre is not None and genre not in distinct_genres:
                    distinct_genres[genre] = 0
            distinct_years = set(years)

            info_tmp = []
            # Yearly entries (total works – all genres)
            for year in distinct_years:
                for genre in distinct_genres:
                    distinct_genres[genre] = 0
                notable_count = 0
                for artist_work in artist_works:
                    release_date = G.nodes[artist_work].get('release_date')
                    genre = G.nodes[artist_work].get('genre')
                    if release_date is not None and int(release_date) == year:
                        if genre:
                            distinct_genres[genre] += 1
                        if G.nodes[artist_work].get('notable'):
                            notable_count += 1
                # Collaborators this year
                collaborator_set = set()
                for collaborator, tuple_set in collaborators.items():
                    for t in tuple_set:
                        if t[1] and int(t[1]) == year:
                            collaborator_set.add(collaborator)
                collaborator_count = len(collaborator_set)
                # Influenced artists this year
                influenced_count = 0
                for influenced, tuple_set in influenced_artists.items():
                    for t in tuple_set:
                        if t[4] and int(t[4]) == year:
                            influenced_count += 1
                info_tmp.append({
                    'year': year,
                    'total_works': years.count(year),
                    'total_notable_works': notable_count,
                    'total_collaborators': collaborator_count,
                    'total_influenced_artists': influenced_count,
                    'total_genres': distinct_genres.copy()
                })

            # Genre entries (total per genre)
            for genre in distinct_genres:
                distinct_genres[genre] = 0
                notable_count = 0
                for artist_work in artist_works:
                    work_genre = G.nodes[artist_work].get('genre')
                    if work_genre == genre:
                        distinct_genres[genre] += 1
                        if G.nodes[artist_work].get('notable'):
                            notable_count += 1
                collaborator_set = set()
                for collaborator, tuple_set in collaborators.items():
                    for t in tuple_set:
                        if t[3] and t[3] == genre:
                            collaborator_set.add(collaborator)
                collaborator_count = len(collaborator_set)
                influenced_count = 0
                for influenced, tuple_set in influenced_artists.items():
                    for t in tuple_set:
                        if len(t) > 7 and t[7] and t[7] == genre:
                            influenced_count += 1
                info_tmp.append({
                    'genre': genre,
                    'total_works': distinct_genres[genre],
                    'total_notable_works': notable_count,
                    'total_collaborators': collaborator_count,
                    'total_influenced_artists': influenced_count
                })

            # Convert oceanus_yearly dict to sorted list of {year, count}
            oceanus_yearly_list = [{'year': yr, 'count': oceanus_yearly[yr]} for yr in sorted(oceanus_yearly.keys())]

            # Append artist record
            data.append({
                'artist_id': node,
                'artist_name': artist_name,
                'total_works': len(artist_works),
                'role_counts': role_counts,
                'oceanus_yearly': oceanus_yearly_list,   # new field
                'info': info_tmp
            })
    return data

def main():
    G = read_data_from_json_to_graph(data_file_path)
    data = prepare_data_for_career_of_artists(G)
    with open("data/artist_career.json", "w") as f:
        json.dump(data, f, indent=2)
    print(f"Saved {len(data)} artists to artist_career.json")

if __name__ == "__main__":
    main()
```
