What genres and top artists have been most influenced by Oceanus Folk?

```python
from helper import *

def find_genres_and_artists_influenced_by(G, genre):
    genres = []
    artists = []
    
    for node, attrs in G.nodes(data=True):
        node_type = attrs.get('Node Type')
        if node_type in ('Song', 'Album'):
            node_genre = attrs.get('genre')
            if node_genre != genre:
                date = attrs.get('release_date')
                for etype in influence_work_types:
                    for neighbor_work in get_neighbors_by_edge_type(G, node, etype , 'out'):
                        neighbor_work_date = G.nodes[neighbor_work].get('release_date')
                        neighbor_genre = G.nodes[neighbor_work].get('genre')
                        if neighbor_genre == genre and int(date) > int(neighbor_work_date):
                            node_genre = attrs.get('genre')
                            genres.append(node_genre)  
                        
                            ## Find artists of this work
                            for ctype in person_work_types:
                                for neighbor_artist in get_neighbors_by_edge_type(G, node, ctype , 'in'):
                                    if G.nodes[neighbor_artist].get('Node Type') == 'MusicalGroup':
                                        for member in get_neighbors_by_edge_type(G, neighbor_artist, member_type, 'in'):
                                            member_name = G.nodes[member].get('name')
                                            artists.append(member_name)
                                    else:
                                        artist_name = G.nodes[neighbor_artist].get('name')
                                        artists.append(artist_name)
    
    return genres, artists

def main():
    G = read_data_from_json_to_graph(data_file_path)
    genres, artists = find_genres_and_artists_influenced_by(G, 'Oceanus Folk')
    distinct_genres = set(genres)
    distinct_artists = set(artists)
    print(len(distinct_artists))
    
    data = {
        'source_genre' : "Oceanus Folk",
        'genres' : [],
        'artists' :[],
    }
    
    for genre in distinct_genres:
        data['genres'].append({
            'name' : genre,
            'count' : genres.count(genre)
        })
    for artist in distinct_artists:
        data['artists'].append({
            'name' : artist,
            'count' : artists.count(artist)
        })
    # Write to json
    with open("data/data_task2-2.json", "w") as f:
        json.dump(data, f, indent=2) 

    # for a in data['artists']:
    #     if a.get('name') == 'Gang Tan':
    #         print(a.get('count'))
    
if __name__ == "__main__":
    main()   
  

```
