```python
from helper import *
from task1_2 import find_collaborators_of_sailor
import json

def generate_collaborator_yearly(G, sailor_node, output_path):
    sailor_collaborators = find_collaborators_of_sailor(G, sailor_node)
    all_years = set()
    collaborator_data = {}

    for collaborator, tuple_set in sailor_collaborators.items():
        collab_years = [int(t[1]) for t in tuple_set]
        first_collab_year = min(collab_years)

        artist_works, _ = find_all_artist_works(G, collaborator)
        oceanus_years = []
        for work in artist_works:
            work_date = G.nodes[work].get('release_date')
            work_genre = G.nodes[work].get('genre')
            if work_genre == 'Oceanus Folk' and work_date:
                year = int(work_date)
                oceanus_years.append(year)
                all_years.add(year)

        if oceanus_years:
            collaborator_data[collaborator] = {
                'name': G.nodes[collaborator].get('name'),
                'firstCollabYear': first_collab_year,
                'years': oceanus_years
            }

    if not collaborator_data:
        print("No collaborators with Oceanus Folk works found.")
        return

    year_min = min(all_years)
    year_max = max(all_years)

    artists_output = []
    for collab, info in collaborator_data.items():
        yearly_counts = {y: 0 for y in range(year_min, year_max + 1)}
        for y in info['years']:
            yearly_counts[y] += 1
        data_points = [{'year': y, 'count': yearly_counts[y]} for y in range(year_min, year_max + 1)]
        artists_output.append({
            'id': collab,
            'name': info['name'],
            'firstCollaborationYear': info['firstCollabYear'],
            'data': data_points
        })

    output_json = {
        'artists': artists_output,
        'yearRange': [year_min, year_max]
    }
    with open(output_path + 'collaborator_yearly.json', 'w') as f:
        json.dump(output_json, f, indent=2)
    print(f"Saved collaborator_yearly.json with {len(artists_output)} collaborators.")

def main():
    G = read_data_from_json_to_graph(data_file_path)
    sailor, _ = find_sailor_shift(G)
    generate_collaborator_yearly(G, sailor, 'data/')

if __name__ == '__main__':
    main()
```
