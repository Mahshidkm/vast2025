```python
from helper import *

import json
from collections import defaultdict

def prepare_oceanus_trend_data(G, sailor_node):
    # Find all Oceanus Folk works (any artist)
    all_oceanus = []
    for node, attrs in G.nodes(data=True):
        if attrs.get('Node Type') in ('Album', 'Song'):
            if attrs.get('genre') == 'Oceanus Folk':
                date = attrs.get('release_date')
                if date and date.isdigit():
                    all_oceanus.append(int(date))
    
    # Find Oceanus Folk works by Sailor
    sailor_works = []
    sailor_notable_works = []
    sailor_works_list, _ = find_all_artist_works(G, sailor_node)  # from helper
    for work in sailor_works_list:
        attrs = G.nodes[work]
        if attrs.get('genre') == 'Oceanus Folk':
            date = attrs.get('release_date')
            if date and date.isdigit():
                year = int(date)
                sailor_works.append(year)
                if attrs.get('notable'):
                    sailor_notable_works.append(year)
    
    # Aggregate counts per year
    years = range(min(all_oceanus), max(all_oceanus)+1)
    all_counts = []
    sailor_counts = []
    notable_counts = []
    
    for y in years:
        all_counts.append(all_oceanus.count(y))
        sailor_counts.append(sailor_works.count(y))
        notable_counts.append(sailor_notable_works.count(y))

    # print(f"all_counts:{all_counts}")
    # print(f"sailor_counts:{sailor_counts}")
    # Build JSON
    data = {
        "years": list(years),
        "all_oceanus": all_counts,
        "sailor_oceanus": sailor_counts,
        "sailor_notable": notable_counts
    }
    with open("data/oceanus_trends.json", "w") as f:
        json.dump(data, f, indent=2)
    print("Saved oceanus_trends.json")

def main():
    G = read_data_from_json_to_graph(data_file_path)
    n, _ = find_sailor_shift(G)
    prepare_oceanus_trend_data(G, n)
    
    
if __name__ == "__main__":
    main()
    
     
```
